```python
import numpy as np
from matplotlib import pyplot as plt
from numpy import genfromtxt
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
%matplotlib inline
```

## Load data

```python
train_file = 'data/problem3data.txt'
train_data = genfromtxt(train_file, delimiter='\t')

y_train = train_data[:, 9]
X_train = train_data[:, :9]
```

# Lasso Code:

```python
w = []
for i in range(0, 9):
    w.append([])
alphas = np.linspace(0.01, 3, 30)

for i in range(0, len(alphas)):
    lasso = Lasso(alpha=alphas[i])
    lasso.fit(X_train, y_train)
    
    for j in range(0, len(lasso.coef_)):
            w[j].append(lasso.coef_[j])
```

```python
plt.figure()
titles = []
for i in range(0, 9):
    plt.plot(alphas, w[i])
    titles.append("w" + str(i + 1))
    
plt.legend(titles)
plt.xlabel('alpha')
plt.ylabel('y')
plt.title('Lasso Regularization')
```

# Ridge Code:

```python
w = []
alphas = []
for i in range(0, 9):
    w.append([])

for i in range(1, 10001):
    alphas.append(i)
    ridge = Ridge(alpha=i)
    ridge.fit(X_train, y_train)
    
    for j in range(0, len(ridge.coef_)):
            w[j].append(ridge.coef_[j])
```

```python
plt.figure()
titles = []
for i in range(0, 9):
    plt.plot(alphas, w[i])
    titles.append("w" + str(i + 1))
plt.legend(titles)
plt.xlabel('alpha')
plt.ylabel('y')
plt.title('Ridge Regularization')
```
